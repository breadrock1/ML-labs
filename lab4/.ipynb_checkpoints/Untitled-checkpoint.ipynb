## Import libs and load dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
import pylab as plt
%matplotlib inline

from matplotlib import pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('/home/akimg/Projects/ML_NeighborsClassifier-master/heart.csv')
data.head(3)
```

```python
X = data[data.columns[:-1]]
y = data['target']
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

## Single tree with no depth limit

```python
tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train)
print("Correctness on training set: {:.3f}".format(tree.score(X_train, y_train)))
print("Correctness on testing set: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file="tree.dot", class_names=["trestbps", "target"], impurity=False, filled=True)
```

```python
import graphviz
with open("tree.dot") as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

## Single tree with limited depth

```python
tree = DecisionTreeClassifier(random_state=0, max_depth=4)
tree.fit(X_train, y_train)
print("Correctness on training set: {:.3f}".format(tree.score(X_train, y_train)))
print("Correctness on testing set: {:.3f}".format(tree.score(X_test, y_test)))
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file="tree.dot", class_names=["trestbps", "target"], impurity=False, filled=True)
```

```python
import graphviz
with open("tree.dot") as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

## Random forest

```python
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier, BaggingRegressor
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
forest = RandomForestClassifier(n_estimators=2, random_state=0)
forest.fit(X_train, y_train)

print("Correctness on training set: {:.3f}".format(forest.score(X_train, y_train)))
print("Correctness on testing set: {:.3f}".format(forest.score(X_test, y_test)))
```

## L1-regularization

```python
from sklearn.linear_model import Ridge, LogisticRegression
for C in [0.001, 1, 100]:
    L1_lr = LogisticRegression(C=C,penalty="l1").fit(X_train, y_train)
    print("L1 train C={:.3f}: {:.4f}".format(C, L1_lr.score(X_train, y_train)))
    print("L1 test  C={:.3f}: {:.4f}".format(C, L1_lr.score(X_test, y_test)))
    print(L1_lr.coef_)
    print()
```

## Cross-validation

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
cross_val_score(estimator=RandomForestClassifier(), X=X_train, y=y_train, cv=5)
```

## Gradient boost

```python
from sklearn.ensemble import GradientBoostingClassifier
```

### Gradient boost

```python
gbrt = GradientBoostingClassifier(random_state=0)
gbrt.fit(X_train, y_train)
print("Correctness on training set: {:.3f}".format(gbrt.score(X_train, y_train)))
print("Correctness on testing set: {:.3f}".format(gbrt.score(X_test, y_test)))
```

### Gradient boost with limited depth

```python
gbrt = GradientBoostingClassifier(random_state=0, max_depth=1)
gbrt.fit(X_train, y_train)
print("Correctness on training set: {:.3f}".format(gbrt.score(X_train, y_train)))
print("Correctness on testing set: {:.3f}".format(gbrt.score(X_test, y_test)))
```

### Gradient boost with a learning speed factor of 0.01

```python
gbrt = GradientBoostingClassifier(learning_rate=0.01)
gbrt.fit(X_train, y_train)
print("Correctness on training set: {:.3f}".format(gbrt.score(X_train, y_train)))
print("Correctness on testing set: {:.3f}".format(gbrt.score(X_test, y_test)))
```

### Gradient boost with a learning speed factor of 1

```python
gbrt = GradientBoostingClassifier(learning_rate=1)
gbrt.fit(X_train, y_train)
print("Correctness on training set: {:.3f}".format(gbrt.score(X_train, y_train)))
print("Correctness on testing set: {:.3f}".format(gbrt.score(X_test, y_test)))
```

### Gradient boost with a learning speed factor of 100

```python
gbrt = GradientBoostingClassifier(learning_rate=100)
gbrt.fit(X_train, y_train)
print("Correctness on training set: {:.3f}".format(gbrt.score(X_train, y_train)))
print("Correctness on testing set: {:.3f}".format(gbrt.score(X_test, y_test)))
```
