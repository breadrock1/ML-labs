```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('C:\\Users\\ЭУНПИ\\Desktop\\student-alcohol-consumption\\student-por.csv', sep=',')
data.head(10)
```

```python
data.info()
```

```python
cols = [0,1,2,3,4,5,8,9,10,11,12,30,31]
data = data.drop(data.columns[cols], axis=1) 
data = data.drop(['Medu'], axis="columns")
data = data.drop(['Dalc'], axis="columns")
```

```python
# Картинку отрисуем с помощью библиотеки seaborn
import seaborn as sns
# Подсчитаем корреляцию и возьмем модуль от нее
corr = data.corr().abs()
# Следующая команда устанавливает размер картинки (по умолчанию она довольно маленькая)
sns.set(rc={'figure.figsize':(12, 14)})
# Нарисуем картинку
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

```python
data.head(2)
```

```python
def yn_to_bin(arg):
    if arg == 'yes':
        return 0
    return 1

data['famsup'] = data['famsup'].apply(yn_to_bin)
data['schoolsup'] = data['schoolsup'].apply(yn_to_bin)
data['nursery'] = data['nursery'].apply(yn_to_bin)
data['higher'] = data['higher'].apply(yn_to_bin)
data['internet'] = data['internet'].apply(yn_to_bin)
data['romantic'] = data['romantic'].apply(yn_to_bin)
data['paid'] = data['paid'].apply(yn_to_bin)
data['activities'] = data['activities'].apply(yn_to_bin)
```

```python
data.head(2)
```

Видим несколько колонок текстовых переменных. Посмотрим, сколько всего таких - выведем типы данных в каждой колонке.

```python
def to_ternary(arg):
    if arg <10:
        return 0
    elif arg <16:
        return 1
    else:
        return 2

data['G3'] = data['G3'].apply(to_ternary)
```

```python
data.head(2)
```

```python
X = data[data.columns[:-1]]
y = data['G3']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
print(X)
```

```python
print(y)
```

```python
# импортируем класс классификатора по ближайшим соседям
from sklearn.neighbors import KNeighborsClassifier

# создадим наш классификатор. Гиперпараметр у этого алгоритма один - количество соседей.
clf = KNeighborsClassifier(n_neighbors=10)
# обучим классификатор на тренировочной выборке
clf.fit(X_train, y_train)
# подсчитаем долю верных ответов алгоритма на тестовой выборке.
# Есть и другие метрики качества - о них будет рассказано на следующей лабе
clf.score(X_test, y_test)
```
